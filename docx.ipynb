```python
doc_filename = os.path.join("docx", "联通手机营业厅使用手册.docx")
doc = zipfile.ZipFile(doc_filename).read("word/document.xml")
root = ET.fromstring(doc)
```

```python
ns = {"w": "http://schemas.openxmlformats.org/wordprocessingml/2006/main"}
body = root.find("w:body", ns)
paragraphs = body.findall("w:p", ns)
for para in paragraphs:
    text_elems = para.findall(".//w:t", ns)
    text = "".join([t.text for t in text_elems])
    print(text)

    # get the style of the paragraph
    try:
        style = para.find("w:pPr", ns).find("w:pStyle", ns)
        # print(style)
        print(style.attrib)
        # style = style.attrib["{http://schemas.openxmlformats.org/wordprocessingml/2006/main}val"]
        # print(style)
    except:
        print("fail")
```

# Split text into tree

```python
import os
import re
from transformers import AutoTokenizer
```

```python
chinese_num = "一二三四五六七八九十"

pattern_list = [
    r"^[%s]\s*[、\.]" % chinese_num, # 一、 一.
    r"^[（〔\(][%s][〕）\)]\s*" % chinese_num, #（一）〔一〕(一) 
    r"^第\s*[%s\d]+\s*部分\s*" % chinese_num, # 第一部分
    r"^\d+\s*[、\)〕）]\s*", # 1、1)
    r"^[（\(〔]\d+[）\)〕]", # （1）(1) 〔1〕
    r"^_*\d+\s*[．\.]\s+", # 1．1. 
    r"^\d(-\d)*\s+", # 1-1, 1-1-1, 1-1-1-1 ... 
    r"^第\s*[%s\d]+\s*章\s*", # 第1章
]
start_pattern = re.compile("|".join(pattern_list))

def match_and_split_heading(text, pattern=start_pattern):
    match = pattern.match(text)
    if match:
        heading_rank = match.group().strip()
        heading_text = text[match.end():].strip()
        return heading_rank, heading_text
    else:
        return None, text
    
def get_heading_info(text):
    for idx in range(len(pattern_list)):
        heading_rank, heading_text = match_and_split_heading(text, re.compile(pattern_list[idx]))
        if heading_rank:
            return idx, heading_rank, heading_text
    return None, None, text

tokenizer = AutoTokenizer.from_pretrained("THUDM/chatglm-6b", trust_remote_code=True)
```

## Split docx with heading

```python
doc_filename = os.path.join("docx", "《中国联通客服知识库》3G业务客服文档.docx")
doc = Document(doc_filename)

# split doc into segments by heading
doc_tree = {}
heading_stack = []
for para in doc.paragraphs:
    if para.style.name.startswith("Heading"):
        level = int(para.style.name[7:].strip())
        heading_stack = heading_stack[:level-1]
        heading_rank, heading_text = match_and_split_heading(para.text)
        if heading_rank is None:
            print("Parse Heading Error: ", para.text)
        heading_stack.append((level, heading_text))
    else:
        if len(heading_stack) > 0 and len(para.text) > 0:
            key = "#".join([x[1] for x in heading_stack])
            if key not in doc_tree:
                doc_tree[key] = ""
            doc_tree[key] += para.text
print(doc_tree)
```

```python
s = "为了集中无线网络的认证功能和策略执行功能。AC 也可以提供集中的桥接、转发"
print(tokenizer.tokenize(s))
print(len(tokenizer.tokenize(s)))
```

## Split docx without heading

```python
MIN_SENTENCE_LEN = 2
MAX_HEADING_LEN = 20

filename = "rfc/tmp_files/RFC5415(中文)无线AP控制和配置CAPWAP协议标准-添加书签版.pdf.txt"
with open(filename, "r", encoding="utf-8") as fin:
    contents = fin.readlines()

doc_tree = {}
heading_stack = []
for line in contents:
    line = line.strip(" \n_- –")
    if len(line) < MIN_SENTENCE_LEN:
        continue
    heading_type, heading_rank, heading_text = get_heading_info(line)
    heading_token_num = len(tokenizer.tokenize(heading_text)) # filter those mistaken headings
    if heading_type is not None and heading_token_num < MAX_HEADING_LEN:
        for idx in range(len(heading_stack)):
            if heading_stack[idx][0] == heading_type:
                heading_stack = heading_stack[:idx]
                break
        heading_stack.append((heading_type, heading_rank, heading_text))
    else:
        if len(heading_stack) > 0:
            key = "#".join([x[2] for x in heading_stack])
            if key not in doc_tree:
                doc_tree[key] = ""
            doc_tree[key] += line

import pprint
pprint.pprint(doc_tree)
```
